```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head(10)
```

## Player Count

* Display the total number of players


```python
#Find total number of players by counting the unique SN of each player
total_players = len(purchase_data_df["SN"].unique())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find number of unique items by counting unique item ID
unique_items = len(purchase_data_df["Item ID"].unique())

#Find the average price
average_price = purchase_data_df["Price"].mean()

#Find the number of purchases
number_purchase = len(purchase_data_df)

#Find the total revenue
total_revenue = purchase_data_df["Price"].sum()

total_purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [average_price], "Number of Purchases": [number_purchase], "Total Revenue": [total_revenue]})
total_purchasing_analysis_df["Average Price"] = total_purchasing_analysis_df["Average Price"].map("${:.2f}".format)
total_purchasing_analysis_df["Total Revenue"] = total_purchasing_analysis_df["Total Revenue"].map("${:,.2f}".format)

total_purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = purchase_data_df.groupby("Gender")
gender_count = gender["SN"].nunique()
percent = gender_count / total_players * 100
percent
gender_demographics = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": percent})
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}%".format)
gender_demographics.index.name = None
gender_demographics = gender_demographics.sort_values("Total Count", ascending=False)
gender_demographics
gender.count().head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = gender["SN"].count()

average_price = gender["Price"].mean()

total_purchase_value = gender["Price"].sum()

average_purchase_gender = total_purchase_value / gender_count

purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": average_price,
                                   "Total Purchase Value": total_purchase_value,
                                   "Avg Total Purchase per Person": average_purchase_gender})

purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

purchasing_analysis.index.name="Gender"

purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names, include_lowest=True)
purchase_data_df.head()

age_group = purchase_data_df.groupby("Age Group")
age_group_count = age_group["SN"].nunique()
age_percentage = age_group_count / total_players * 100

age_demographics = pd.DataFrame({"Total Count": age_group_count,
                                "Percentage of Players": age_percentage})
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:.2f}%".format)
age_demographics.index.name = None
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_count = age_group["SN"].count()

age_avg_price = age_group["Price"].mean()

age_total_value = age_group["Price"].sum()

average_purchase_age = age_total_value / age_group_count

age_purchasing_analysis = pd.DataFrame({"Purchase Count": age_purchase_count,
                                     "Average Purchase Price": age_avg_price,
                                     "Total Purchase Value": age_total_value,
                                     "Avg Total Purchase per Person": average_purchase_age})

age_purchasing_analysis["Average Purchase Price"] = age_purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
age_purchasing_analysis["Total Purchase Value"] = age_purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_purchasing_analysis["Avg Total Purchase per Person"] = age_purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)

age_purchasing_analysis.index.name="Age Ranges"

age_purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data_df.groupby("SN")

top_purchase_count = top_spenders["SN"].count()

top_avg_purchase_price = top_spenders["Price"].mean()

top_total_purchase = top_spenders["Price"].sum()

top_spender_analysis = pd.DataFrame({"Purchase Count": top_purchase_count,
                                    "Average Purchase Price": top_avg_purchase_price,
                                    "Total Purchase Value": top_total_purchase})

top_spender_analysis = top_spender_analysis.sort_values("Total Purchase Value", ascending=False)

top_spender_analysis["Average Purchase Price"] = top_spender_analysis["Average Purchase Price"].map("${:.2f}".format)
top_spender_analysis["Total Purchase Value"] = top_spender_analysis["Total Purchase Value"].map("${:.2f}".format)

top_spender_analysis.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular = purchase_data_df[["Item ID", "Item Name", "Price"]]

items = popular.groupby(["Item ID", "Item Name"])

item_count = items["Item Name"].count()

item_total_purchase = items["Price"].sum()

item_price = item_total_purchase / item_count

item_analysis = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": item_total_purchase})

sorted_item_analysis = item_analysis.sort_values("Purchase Count", ascending=False)

sorted_item_analysis["Item Price"] = sorted_item_analysis["Item Price"].map("${:.2f}".format)
sorted_item_analysis["Total Purchase Value"] = sorted_item_analysis["Total Purchase Value"].map("${:.2f}".format)

sorted_item_analysis.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable = item_analysis.sort_values("Total Purchase Value", ascending=False)

most_profitable["Item Price"] = most_profitable["Item Price"].map("${:.2f}".format)
most_profitable["Total Purchase Value"] = most_profitable["Total Purchase Value"].map("${:.2f}".format)

most_profitable.head()
```
